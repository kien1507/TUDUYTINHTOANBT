## W3A2
Hoán đổi hai số không sd biến tạm thời
Sử dụng phép toán XOR bitwise để hoán đổi giá trị của hai biến
and (&): and bit (1 & 1 = 1 , còn lại = 0)
or (|): or bit (0 | 0 = 0 , còn lại = 0)
xor (^): xor bit (khác nhau bằng 1, giống nhau bằng 0)
not (!): lật bit (1 => 0, 0 => 1)
thập phân => nhị phân: chia 2 lấy dư đến kết quả 0, lấy ngược dưới lên 

```python
a, b = map(int, input().split())
print(f"a = {a}, b = {b}")
a = a ^ b # thay thế 3 dong bằng a, b = b, a
b = a ^ b
a = a ^ b
print(f"hoán đổi a = {a}, b = {b}")
```

## W3A3

Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

```python
n = int(input())
if n > 0 and (n & (n-1) == 0):
    print(f"{n} là luy thưa cua 2")
else :
    print(f"{n} ko la luy thua cua 2")
```

## W3A11 [Phân loại tam giác]
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường". 


```python
a, b, c = map(int, input().split())
if a+b>c and a+c>b and b+c>a:
    if a == b == c:
        print("tam giác đều")
    elif a == b or b == c or a == c:
        print ("tam giác cân")
    else:
        print("tam giác thường")
else:
    print("không phải tam giác")    
```

## W3A13 [Tính tiền điện]
Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh 


```python
n = int(input())
if n >= 0 and n <= 50:
    tien = n*1500
elif n >= 51 and n <= 100:
    tien = 50*1500 + (n-50)*2000
elif n > 101 :
    tien =  50*1500 + 50*2000 + (n-100)*3000
print(tien)
```

## W3A16 [Rounding]
Nhập vào một số thực: in ra kết quả làm tròn lên ,tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm
sẵn có của pyton). Các kết quả cách nhau một dấu cách

```python
n = float(input())
x = n//1        # làm tròn xuống
y = n//1 + 11   # làm tròn lên
if (n*10)%10 >= 5:
    print(f"{y} {x} {y}")
else:
    print(f"{y} {x} {x}")

```

## W3A17 Cấp số nhân
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng
thứ tự a, b, c, d hay không?
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội và so sánh vs c, c nhân công bội và so sánh vs
d. 


```python
a, b, c, d = map(float, input().split())
n = b/a
if b*n==c:
    if c*n==d:
        print('la cap so nhan')
else:
    print("ko phai cap so nhan")
```
