## Bài 4: 
Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím
Ví dụ: Ta nhập h = 4:
   *
  * *
 *   *
******* 

```python
h = int(input("Nhập độ cao h: "))
for i in range(1, h + 1):
    khoang_trang_ngoai = " " * (h - i)
    if i == 1:
        print(khoang_trang_ngoai + "*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        khoang_trang_trong = " " * (2 * i - 3)
        print(khoang_trang_ngoai + "*" + khoang_trang_trong + "*")
```

## Bài 12:

lớn nhất nhưng nhỏ hơn n 


```python
n = int(input())
s = 0
k = 1
while s + k < n:
    k += 1
    s += k
print(k)
```

## Bài 13:

1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

## Bài 14: 
Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
dem = 0
while True:
    n = int(input())
    if n == -1:
        break
    dem += 1
    if dem==1:
        min = n
        max = n
    else:
        if n > max:
            max = n
        if n < min:
            min = n
print(max,min)
```

## Bài 15: 
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều
cách nhất có thể, ko dùng mảng. 


```python
n = int(input())
dem = 0
if n == 0:
    dem == 0
else:
    while n > 0:
        n //= 10
        dem += 1
print(dem)
```

## Bài 16: 
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ 


```python
n = int(input())
chan = 0
le = 0
if n == 0:
    chan = 1
else:
    while n > 0:
        k = n % 10
        if k % 2 == 1:
            le += 1
        else:
            chan += 1
        n = n // 10
print(le,chan)  
```

Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay
không 


```python
n = int(input())
if n <= 0:
    print("khong co dang 3^k")
else:
    while n % 3 == 0:
        n //= 3
        if n == 1:
            print("có dạng 3^k")
    else:
        print("ko có dạng 3^k")
```

## Bài 19: 
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 


```python
a = int(input())
b1 = b2 =1
if a < 1:
    print("khong co so Fibonacci thoa man")
elif a == 1:
    print("so Fibonacci thoa man la 1")    
else:
    b3 = b1 + b2
    while b3 <= a:
        b1 = b2
        b2 = b3
        b3 = b1 + b2
    print("so Fibomnacci thoa man la", b2)
```
