## W5A16
Viết hàm kiểm tra một số có phải số Amstrong hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_armstrong(x):
    s = str(x)
    n = len(s)
    tong = 0
    for chu_so in s:
        tong += int(chu_so) ** n
    if tong == x:
        return True
    else:
        return False
is_armstrong(153)
    
```

## W5A19 
VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.


```python
def la_so_ky_dieu(n):
    if n <= 0:
        return False
        
    while n > 9:
        tong = 0
        while n > 0:
            tong += n % 10
            n //= 10
        n = tong 
        
    return n == 1
 la_so_ky_dieu(123)
```

Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

## W5A27 
Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count_non_zero_digits(lst):
    dem = 0
    for x in lst:
        s = str(abs(x))
        for char in s:
            if char != '0':
                dem += 1
    return dem
lst = [1,0,3,30]
count_non_zero_digits(lst)
```

## W5A28 
Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
def tinh_goc_abc(A, B, C):
    x1, y1 = A
    x2, y2 = B
    x3, y3 = C
    ba_x, ba_y = x1 - x2, y1 - y2
    bc_x, bc_y = x3 - x2, y3 - y2
    dot_product = ba_x * bc_x + ba_y * bc_y
    mag_ba = (ba_x**2 + ba_y**2)
    mag_bc = (bc_x**2 + bc_y**2)
    if mag_ba == 0 or mag_bc == 0:
        return 0
    cosine_angle = dot_product / (mag_ba * mag_bc)
    cosine_angle = max(-1.0, min(1.0, cosine_angle))
    angle_rad = math.acos(cosine_angle)
    angle_deg = math.degrees(angle_rad)

    return angle_deg
A = (0, 1)
B = (0, 0)
C = (2, 3)
tinh_goc_abc(A, B, C)
```

## W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 

```python
def has_unique_digits(x):
    s = str(x)
    return len(s) == len(set(s))

def sum_1_to_n(n):
    t = 0
    for x in range(1, n + 1):
        if has_unique_digits(x):
            t += x
    return t
print(sum_1_to_n(15))
```
