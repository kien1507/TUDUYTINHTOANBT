### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không

```python
def a(lst):
    n = len(lst)
    for i in range(n//2):
        if lst[i]!= lst[n-1-i]:
            return False
    else:
        return True
 
lst = list(map(int, input().split()))
a(lst)
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
def dich_khong(lst):
    for i in range(len(lst)):
        if lst[i] == 0:
            k = lst.pop(i)
            lst.insert(0,k)
    return lst
lst = list(map(int, input().split()))
dich_khong(lst)
    
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
def check_ds(lst):
    if len(lst) < 2:
        return "Vo danh"
    
    tang = False
    giam = False
    bang = False

    for i in range(len(lst) - 1):
        if lst[i] < lst[i+1]:
            tang = True
        elif lst[i] > lst[i+1]:
            giam = True
        else:
            bang = True
    if tang and not giam and not bang:
        return "Tang"
    if giam and not tang and not bang:
        return "Giam"
    return "Vo danh"
lst = list(map(int, input().split()))
print(check_ds(lst))
```

### Bài 8: Kiểm tra một list số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac

```python
def check_ziczac(lst):
    if len(lst) < 3:
        return "khong phai ziczac"
    for i in range(1, len(lst) - 1):
        if not ((lst[i-1] < lst[i] > lst[i+1]) or (lst[i-1] > lst[i] < lst[i+1])):
            return "Khong phai ziczac"
    else:
        return "ziczac"
lst = list(map(int, input().split()))
check_ziczac(lst)
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về "-1"

```python
def in_vitri(k,lst):
    vtri = []
    xuathien = False
    for i in range(len(lst)):
        if lst[i] == k:
            vtri.append(i)
            xuathien = True
    if xuathien:
        return vtri
    else:  
        return "-1"
n = list(map(int, input().split()))
in_vitri(3,n)
```
