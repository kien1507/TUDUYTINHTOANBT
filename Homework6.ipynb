## W6A16
Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n = int(input())
m = int(input())
matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    if len(row) == m:
        matrix.append(row)
for row in matrix:
    for x in row:
        print("%4d" % x, end="") 
    print()
```

## W6A18
Nhập vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột k (hàng k)

```python
def tong_pt(n,m,k):
    matrix = []
    for i in range(n):
        row = list(map(int, input().split()))
        if len(row) == m:
            matrix.append(row)
    tong_cotk = 0
    for i in range(len(matrix)):
        tong_cotk += matrix[i][k-1]
    tong_hangk = 0
    for i in range(len(matrix[k-1])):
        tong_hangk += matrix[k-1][i]
    return tong_cotk,tong_hangk
n, m, k = map(int, input().split())
tong_pt(n,m,k)
    
```

## W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
def anh_xa_vi_tri(s):
    lst_s = s.split()
    d = {}
    for i, chuoi in enumerate(lst_s):
        if chuoi in d:
            d[chuoi].append(i)
        else:
            d[chuoi] = [i]
    for chuoi in d:
        d[chuoi] = tuple(d[chuoi])
    return d
s = input()
anh_xa_vi_tri(s)
```

## W6A23
VIết hàm xóa đi các số < x có trong mảng

```python
# CACH 1
def xoa_phantu_theo_dk(lst,x):
    for i in range(len(lst) - 1, -1, -1):
        if lst[i] < x:
            lst.pop(i)
    return lst
lst = list(map(int, input().split()))
x = int(input())
xoa_phantu_theo_dk(lst,x)
```

```python
# CACH 2
def xoa_phantu_theo_dk(lst,x):
    kq = []
    for i in lst:
        if i >= x:
            kq.append(i)
    return kq
lst = list(map(int, input().split()))
x = int(input())
xoa_phantu_theo_dk(lst,x)
```

## W6A24
Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def đẩy_số_0(lst):
    day_0 = []
    day_khac = []
    for i in lst:
        if i == 0:
            day_0.append(i)
        else:
            day_khac.append(i)
    return day_0 + day_khac
lst = list(map(int, input().split()))
print(*đẩy_số_0(lst))
```

## W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối

```python
def tong_cuc_dai_dia_phuong(lst):
    tong = 0
    n = len(lst)
    for i in range(1, n - 1):
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            tong += lst[i]
    return tong
lst = list(map(int, input().split()))
tong_cuc_dai_dia_phuong(lst)
```

## W6A27
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước) nếu không có in ra -1

```python
lst = list(map(int, input().split()))
def vtri_so_can_tim(lst,x):
    kq = []
    for i,v in enumerate(lst):
        if v < x:
            kq.append(i)
    if len(kq) == 0:
        kq = [-1]
    return kq
vtri_so_can_tim(lst,3)
```

## W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm thực sự, hay vô danh (không tăng, không giảm)

```python
def check_ds(lst):
    if len(lst) < 2:
        return "Vo danh"
    tang = False
    giam = False
    bang = False
    for i in range(len(lst) - 1):
        if lst[i] < lst[i+1]:
            tang = True
        elif lst[i] > lst[i+1]:
            giam = True
        else:
            bang = True
    if tang and not giam and not bang:
        return "Tang"
    if giam and not tang and not bang:
        return "Giam"
    return "Vo danh"
lst = list(map(int, input().split()))
print(check_ds(lst))
```
