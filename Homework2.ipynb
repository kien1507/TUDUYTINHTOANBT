## Bài 3 :
 Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a, b, c = map(int, input().split())
p = (a+b+c)/2
if a+b>c and a+c>b and b+c>a:
    if a == b == c:
        print("tam giác đều")
    elif a == b or b == c or a == c:
        print ("tam giác cân")
    else:
        print("tam giác thường")
    print(f"dien tich {(p*(p-a)*(p-b)*(p-c))**0.5}, chu vi {2*p}")
else:
    print("không phải tam giác") 
```

## Bài 5:
 Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(float, input().split())
delta = b**2 - 4*a*c
if delta > 0:
    x1 = (-b + delta**0.5)/(2*a)
    x2 = (-b - delta**0.5)/(2*a)
    print(f"co 2 nghiem phan biet {x1}, {x2}")
elif delta == 0:
    x = -b/(2*a)
    print(f"co nghiem kep {x}")
else:
    print("vo nghiem")
        
```

## Bài 6:
 Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bànphím).

## Bài 8: Giải hệ phương trình tuyến tính:
ax+by=m
cx+dy=n

```python
a, b ,m ,c, d, n = map(float, input().split())
D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c
if D != 0:
    x = Dx / D
    y = Dy / D
    print(f"Hệ phương trình có nghiệm duy nhất: x = {x}, y = {y}")
else:
    if Dx == 0 and Dy == 0:
        print("Hệ phương trình có vô số nghiệm.")
    else:
        print("Hệ phương trình vô nghiệm.")
```

## Bài 9:
 Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây

```python
x = int(input())
h = x // 3600
m = x % 3600 // 60
s = x % 60
print(f"{x} giây = {h} giờ {m} phút {s} giây")
```

## Bài 10:
 Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
a, b, x, y, r = map(float, input().split())
d = ((x-a)**2 + (y-b)**2)**0.5
if d == r :
    print(f"điểm A ({x},{y}) thuộc đường tròn tâm ({a}, {b}) bán kính {r}")
else:
    print(f"điểm A ({x},{y}) không thuộc đường tròn tâm ({a}, {b}) bán kính {r}")
```
