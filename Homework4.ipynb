W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n = int(input())
if n == 0 :
    print("co dang 2^k")
else:
    while n % 2 == 0:
        n //= 2
        if n == 1:
            print("co dang 2^k")
        else:
            print("ko co dang 2^k")

```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
dem = 0
lonnhat = nhonhat = 0
while True:
    n = int(input())
    if n == -1:
        break
    dem += 1
    if dem == 1:
        lonnhat = nhonhat = n
    else:
        if n > lonnhat:
            lonnhat = n
        if n < nhonhat:
            nhonhat = n
print(lonnhat,nhonhat)
```

W4A29
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
thường và thường thành hoa

```python
s = input("Nhập chuỗi: ")
ket_qua = ""
for i in range(len(s)):
    char = s[i]
    if 65 <= ord(char) <= 90:
        ket_qua += chr(ord(char) + 32)
    elif 97 <= ord(char) <= 122:
        ket_qua += chr(ord(char) - 32)
    else:
        ket_qua += char

print(ket_qua)
```
